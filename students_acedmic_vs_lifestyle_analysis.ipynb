# **Data Visualisation**

## Univariate Analysis

### 1. Age Distribution of Students

```python
sns.countplot(x = 'age', data = df, width = 0.5)
plt.title("Age Distribution of Students")
plt.show()
```

### 2. Gender Distribution

```python
sizes = [52.3, 47.7]
df['gender'].value_counts().plot(kind = 'pie', autopct = '%1.2f%%', startangle = 90,explode = [0.1, 0] if sizes[0] > sizes[1] else [0, 0.1],shadow=True)
plt.ylabel('')
plt.title('Gender Distribution')
plt.legend(title='Categories',loc='center left',bbox_to_anchor=(1,1))
plt.show()
```

### 3. Sleep Hours per Day

```python
sns.histplot(df['sleep_hours'], bins = 10, kde = True, color = 'b')
plt.title('Sleep Hours per Day')
plt.xlabel('Sleep Hours')
plt.ylabel('count')
plt.show()
```

### 4. Part-Time Job Status

```python
value_order = df['part_time_job'].value_counts()
labels = value_order.index.tolist()

color_map = {'Yes': 'blue', 'No': 'red'}
colors = [color_map[label] for label in labels]

value_order.plot(kind = 'pie', autopct = '%1.2f%%', startangle = 90, colors = colors)
plt.title('Part-Time Job Status')
plt.ylabel('')
plt.show()
```

### 5. Diet Quality

```python
diet_order = df['diet_quality'].value_counts()
labels = diet_order.index.tolist()

color_map = {'Good':'green', 'Fair': 'yellow', 'Poor': 'red'}
colors = [color_map[label] for label in labels]

diet_order.plot(kind = 'pie', autopct = '%1.2f%%', startangle = 90, colors = colors)
plt.title('Diet Quality')
plt.ylabel('')
plt.show()
```

### 6. Exercise Frequency per Day

```python
sns.countplot(x = 'exercise_frequency', data = df)
plt.title('Exercise Frequency per Day')
plt.xlabel('Days')
plt.ylabel('Count')
plt.show()
```

### 7. Parental Education Level Distribution

```python
labels = df['parental_education_level'].value_counts().index

color_map = {'Master': 'green','Bachelor': 'yellow','High School': 'red'}
colors = [color_map[label] for label in labels]

plt.figure(figsize=(10, 6)) 

df['parental_education_level'].value_counts().plot.pie(autopct='%1.1f%%',startangle=90,colors=colors,
        explode=[0.05 if val == max(df['parental_education_level'].value_counts()) else 0 for val in df['parental_education_level'].value_counts()],shadow=True)

plt.title('Parental Education Level Distribution', fontsize=16)
plt.ylabel('')
plt.legend(title='Categories', loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```

### 8. Mental Health Rating

```python
sns.histplot(df['mental_health_rating'], bins=10, kde=True, color = 'r')
plt.title("Mental Health Rating")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

### 9. Extracurricular Participation

```python
Extracurricular_order = df['extracurricular_participation'].value_counts()
labels = Extracurricular_order.index.tolist()

color_map = {'Yes': 'blue', 'No': 'red'}
colors = [color_map[label] for label in labels] 

Extracurricular_order.plot(kind = 'pie', autopct = '%1.2f%%', startangle = 90, colors = colors)
plt.title('Extracurricular Participation')
plt.ylabel('')
plt.show()
```

### 10. Subplotting: 
####     (i) Distribution of Study Hours
####     (ii) Distribution of Hours Spent on Screen Everyday

```python
plt.figure(figsize = (15,6))
#plotting for study hours per day
plt.subplot(1, 2, 1)#(no. of rows, no. of columns, position of subplot)
plt.hist(df['study_hours_per_day'], width = 0.8,color = 'purple')
plt.xlabel('Study Hours')
plt.ylabel('Count')
plt.title('Distribution of Study Hours')

#plotting for screen time per day
plt.subplot(1, 2, 2)#(no. of rows, no. of columns, position of subplot)
plt.hist(df['screen_time_per_day'], width = 0.95,color = 'magenta')
plt.xlabel('Screen Time (hrs)')
plt.ylabel('Count')
plt.title('Distribution of Hours Spent on Screen Everyday')

plt.show()
```

## Bivariate Analysis

### 1. Study Hours vs Exam Score

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize = (10, 6))
sns.scatterplot(x = 'study_hours_per_day', y = 'exam_score', data = df)
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours per Day')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```

### 2. Average Exam Score vs Sleep Hours

```python
sleep_vs_score = df.groupby('sleep_hours')['exam_score'].mean().reset_index()

plt.figure(figsize = (10,6))
sns.lineplot(x = 'sleep_hours', y = 'exam_score', data = sleep_vs_score, marker = 'o', linewidth = 2)
plt.title('Average Exam Score vs Sleep Hours')
plt.xlabel('Sleep Hours per Day')
plt.ylabel('Average Exam Score')
plt.grid(True)
plt.show()
```

### 3. Subplot: 
####   (i) Screen Time vs Exam Score (Scatter Plot)
####   (ii) Screen Time vs Exam Score (Line Plot)

```python
plt.figure(figsize=(16, 6))

# Scatter plot
plt.subplot(1, 2, 1)
sns.scatterplot(x='screen_time_per_day', y='exam_score', data=df)
plt.xlabel('Screen Time per Day (hrs)')
plt.ylabel('Exam Score')
plt.title('Scatter Plot: Screen Time vs Exam Score')
plt.grid(True)

# Line plot
plt.subplot(1, 2, 2)
sns.lineplot(x='screen_time_per_day', y='exam_score', data=df, marker='o', color='orange')
plt.xlabel('Screen Time per Day (hrs)')
plt.ylabel('Exam Score')
plt.title('Line Plot: Screen Time vs Exam Score')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Multivariate Analysis

### 1. Study Hours vs Exam Score by Gender
